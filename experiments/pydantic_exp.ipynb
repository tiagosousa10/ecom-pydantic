```python
from pydantic import BaseModel

class Player(BaseModel):
    name: str
    score: int
```

```python
player1 =Player(name="John", score=100)
print("Player 1:", player1)
print("-------")
```

```python
player2 = Player(name="Jane", score="200")
print("Player 2:", player2)
print("-------")
```

```python
player3= Player(name="John", score="Fifty")
print("Player 3:", player3)
print("-------")
```

```python
basic_agent = Agent("groq:llama-3.1-8b-instant")

response = await basic_agent.run(
    "Why the sky is blue?, give a reply in a funny way")


print("Response:", response.output)
```

```python
pirate_agent=Agent(
    name="monkey-d-luffy",
    description="a great captain of the pirate ship",
    model='groq:llama-3.1-8b-instant',
    system_prompt=(
        "You are a pirate. You will be asked questions and you will answer them in pirate language.",
        "You are the captain of the ship. You will be asked questions and you will answer them in pirate language.",
        "Never drop your character. You will be asked questions and you will answer them in pirate language.",
    )
)
```

```python
pirate_agent
```

```python
response = await pirate_agent.run(
    "Explain how to get food on a ship. anwser in one sentence"
)

print("Response:", response.output)
```
